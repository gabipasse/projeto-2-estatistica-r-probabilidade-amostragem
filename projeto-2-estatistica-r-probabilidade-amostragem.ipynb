# **Em um concurso para preencher uma vaga de cientista de dados temos um total de 10 questões de múltipla escolha com 3 alternativas possíveis em cada questão. Cada questão tem o mesmo valor. Suponha que um candidato resolva se aventurar sem ter estudado absolutamente nada. Ele resolve fazer a prova de olhos vendados e chutar todas as resposta. Assumindo que a prova vale 10 pontos e a nota de corte seja 5, obtenha a probabilidade deste candidato acertar 5 questões e também a probabilidade deste candidato passar para a próxima etapa do processo seletivo.**

**Qual o número de ensaios(n)?**

```r
n <- 10

n
```

**Os ensaios sao independentes?**

Sim. A opção escolhida em uma questão não influencia nas opções escolhidas nas demais questões

**Somente dois resultados são possíveis em cada ensaio?**

Sim. O candidato tem duas possibilidades, ERRAR ou ACERTAR uma questão

**Probabilidade de sucesso(p)**

```r
p <- 1 / 3

p
```

**Probabilidade de fracasso(q)**

```r
q <- 1 - p

q
```

**Qual o total de eventos que deseja obter sucesso(k)**

```r
k <- 5

k
```

**Solução 1:**

```r
probabilidade <- choose(n, k) * (p ** k) * (q ** (n - k))

probabilidade
```

**Solução 2:**

```r
dbinom(x = k, size = n, prob = p)
```

**Probabilidade do candidato passar (pelo menos 5 acertos)**

P(acertar >= 5) = P(5) + P(6) + P(7) + P(8) + P(9) + P(10)

```r
probabilidade <- 0

for (i in 5:10) {
    probabilidade <- probabilidade + dbinom(x = i, size = n, prob = p)
}

probabilidade
```

```r
probabilidade <- sum(dbinom(x = 5:10, size = n, prob = p))

probabilidade
```

```r
probabilidade <- pbinom(q = 4, size = n, prob = p, lower.tail = FALSE)

probabilidade
```

O fabricante estipula um erro máximo sobre a média populacional de apenas 5 ml. Para garantir um nível de confiança de 95% qual o tamanho de amostra deve ser selecionado para este estudo?**

```r
determinando_tamanho_amostra <- function(Z, s, N, e){
    
    n <- ((Z ** 2) * (s ** 2) * N) / (((Z ** 2) * ( s ** 2)) + (e ** 2) * (N - 1) )

    return(round(n))
    
}
```

```r
set.seed(100)

amostra_requisito_1 <- sample_n(dados, 200)$Renda

head(amostra_requisito_1)
```

```r
media_amostra_req_1 <- mean(amostra_requisito_1)

media_amostra_req_1
```

```r
sd_amostra_req_1 <- sd(amostra_requisito_1)

sd_amostra_req_1
```

**2**

```r
e <- 0.1 * media_amostra_req_1

e
```

```r
# definir funcao para pegar Z a partir do nivel de confianca
```

```r
determinando_Z <- function(confianca){
    
    probabilidade <- confianca / 2 + 0.5
    
    print(probabilidade)
    
    return(qnorm(probabilidade))
    
}
```

```r
determinando_tamanho_amostra_pop_infinita <- function(Z, e, s){
    
    n <- ( Z * s / e) ** 2
    
    return(n)
    
}
```

```r
valores_n <- c()
```

```r
for (confianca in c(0.9, 0.95, 0.99)) {
    
    Z <- determinando_Z(confianca)
    
    print(Z)
    
    valores_n <- c(valores_n, determinando_tamanho_amostra_pop_infinita(Z, e, sd_amostra_req_1))
    
}
```

```r
valores_n
```

**3**

```r
custos_pesquisa <- c()
```

```r
custo_entrevista <- 100

custo_entrevista
```

```r
custos_pesquisa <- valores_n * custo_entrevista
```

```r
custos_pesquisa
```
